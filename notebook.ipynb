2. Supervised Learning - Regression

Target Variable: TotalSpending (continuous)

• Train/Test Split: 80/20 split, random_state=42.

• Models to Implement (using scikit-learn):

  - Linear Regression
  - Ridge Regression (alpha=1.0)
  - Decision Tree Regressor (max_depth=5)

• Evaluation Metrics: MSE, RMSE, R2 Score.

• Feature Scaling: Apply StandardScaler to numerical features (excluding target).

• Output: A comparison table of all three regression models showing their performance.

```python
print(df.columns)
```

```python
#  delete the data column, it's not useful for the model training   
df = df.drop(columns=['Dt_Customer'])
```

### Train / Test Split

```python
from sklearn.model_selection import train_test_split

#  drop the target form the datast 
x = df.drop(columns=['TotalSpending'])
# define the target variable
y = df['TotalSpending']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

### Feature Scaling


```python
from sklearn.preprocessing import StandardScaler

# numerical features to scale
num_features = [
    'Income', 'Age', 'TotalChildren',
    'NumWebPurchases', 'NumStorePurchases',
    'NumCatalogPurchases', 'NumDealsPurchases',
    'Recency'
]

#  train the data scaler on the training data and transform both training and test data
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train[num_features])
x_test_scaled = scaler.transform(x_test[num_features])

```

```python
# replace the original numerical features with the scaled ones
x_train[num_features] = x_train_scaled
x_test[num_features] = x_test_scaled
```

### build models and train the models 


#### 1. Linear Regression

* **Definition:** A baseline regression model that assumes a straight-line (linear) relationship between the independent features and the target variable (`TotalSpending`).
* **Purpose:** Used as a reference point to evaluate if the data has simple linear patterns and to compare against more complex models.



#### 2. Ridge Regression

* **Definition:** An extension of linear regression that applies **L2 Regularization**, adding a penalty to prevent feature weights from becoming too large.
* **Purpose:** Used to prevent **Overfitting** and handle highly correlated features (Multicollinearity), ensuring more stable and reliable predictions.

#### 3. Decision Tree Regressor

* **Definition:** A non-linear model that splits data into smaller subsets based on a series of nested yes/no questions, forming a tree-like structure.
* **Purpose:** Used to capture complex, non-linear interactions between features (like age, income, and children) that linear models fail to see.

```python
# import labraries for models
from sklearn.linear_model import Ridge
from sklearn.linear_model import LinearRegression

from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

```

```python
# build and train the models 

# define the required modls 
models = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(alpha=1.0),
    'Decision Tree Regressor': DecisionTreeRegressor(max_depth=5, random_state=42)
}

# store the preformmance results 
performance_results = {}

#  for loop for training and evalustion  

for name, model in models.items():
    # train the model on the scaled training data 
    model.fit(x_train_scaled, y_train)
    
    # predict  the test data
    y_pred = model.predict(x_test_scaled)
    
    # claculate performance metrics
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    
    # save the results 
    performance_results[name] = {
        'MSE': mse,
        'RMSE': rmse,
        'R2 Score': r2
    }
```

```python
# preformance table 
df_comparison = pd.DataFrame(performance_results).T

print("Regression Models PERFOMANCE COMPARISON:")
print(df_comparison.round(4))

```

```python
x_train_c_scaled = scaler.fit_transform(x_train_c)
x_test_c_scaled = scaler.transform(x_test_c)
```

#### Build , Train and Evaluate the Models


**1. Logistic Regression**

* **Definition:** A fundamental classification model that uses a logistic (sigmoid) function to model the probability of a binary outcome (0 or 1).
* **Purpose:** Used with `class_weight='balanced'` to establish a probabilistic baseline while directly addressing the heavy class imbalance in customer responses.



**2. K-Nearest Neighbors (KNN)**

* **Definition:** A distance-based classifier that predicts a data point's class based on the majority vote of its $k$ closest neighbors in the feature space.
* **Purpose:** Used to capture local data patterns and similarity clusters among customers, strictly requiring feature scaling to calculate distances accurately.



**3. Random Forest Classifier**

* **Definition:** An ensemble learning model that constructs a multitude of independent decision trees at training time and outputs the majority vote of the individual trees.
* **Purpose:** Used to capture complex, non-linear interactions and handle class imbalance robustly without requiring any feature scaling.

```python


from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

```python
#  define the models in a tuple 
models_clf = {
#  logictics  regression with scaled trained  data 
    'Logistic Regression': (LogisticRegression(class_weight='balanced', random_state=42), x_train_c_scaled, x_test_c_scaled),
    #  KNN with scaled trained  data 
    'K-Nearest Neighbors': (KNeighborsClassifier(n_neighbors=5), x_train_c_scaled, x_test_c_scaled),
    #  Random Forest with unscaled trained  data 
    'Random Forest Classifier': (RandomForestClassifier(n_estimators=100, random_state=42), x_train_c, x_test_c) 
}
```

```python
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
clf_results = {}  # for storing classification results

for name, (model, train_data, test_data) in models_clf.items():
  
    print(f" * Model: {name}")
  
    
    # train and predict 
    model.fit(train_data, y_train_c)
    y_pred_c = model.predict(test_data)
    
    # Accuracy, Precision, Recall, F1
    acc = accuracy_score(y_test_c, y_pred_c)
    prec = precision_score(y_test_c, y_pred_c, pos_label=1)
    rec = recall_score(y_test_c, y_pred_c, pos_label=1)
    f1 = f1_score(y_test_c, y_pred_c, pos_label=1)
    
    # store results 
    clf_results[name] = {
        'Accuracy': acc,
        'Precision (Class 1)': prec,
        'Recall (Class 1)': rec,
        'F1-Score (Class 1)': f1
    }
    
    # Calculate and print confusion matrix
    cm = confusion_matrix(y_test_c, y_pred_c)
    print("Confusion Matrix:")
    print(cm)
    
    # Plot confusion matrix in a visually appealing way
    plt.figure(figsize=(4,3))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['No', 'Yes'], yticklabels=['No', 'Yes'])
    plt.title(f'Confusion Matrix - {name}')
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()


df_clf_comparison = pd.DataFrame(clf_results).T
print("Classification Models Performance Metrics:")
print(df_clf_comparison.round(4))
    
```

## 4. Unsupervised Learning - Clustering



Scaling → Elbow → Fit → Visualization → Profiling  

Goal: Segment customers using TotalSpending, Income, Age, and TotalChildren.

• Data Preparation: Select the 4 features above, scale them using StandardScaler.

• K-Means Clustering:
- Use the Elbow Method (inertia vs. k from 1 to 10) to determine optimal k.
- Fit K-Means with the chosen k (random_state=42).
- Add cluster labels to the original dataframe.


• Visualization: Create a 2D scatter plot (using PCA reduction to 2 components) colored by
cluster label.


• Cluster Profiling: For each cluster, compute mean values of Income, TotalSpending, Age, TotalChildren. 

Briefly describe each customer segment (e.g., "High-income, high-spending young singles").


#### Data Preparation and scaling 

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
# prepare the selected 4 features 
cluster_features = ['TotalSpending', 'Income', 'Age', 'TotalChildren']

x_cluster = df[cluster_features]
#  apply scaling to the selected features for clustering
x_cluster_scaled = StandardScaler().fit_transform(x_cluster)
```

#### K-Means Clustering:

```python
# Elbow method to find optimal number of clusters

inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(x_cluster_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia, marker='o')

plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.xticks(range(1, 11))
plt.show()
```

omptimal k  is 3 ; based on the elbow method 

```python
# fit the model with the optimal k 
kmeans_optimal = KMeans(n_clusters=3, random_state=42, n_init=10)
#  train the model on scaled data 
kmeans_optimal.fit(x_cluster_scaled)


# assign cluster labels to the original dataframe
df['Cluster'] = kmeans_optimal.labels_
df.head()
```

####  Visualization with PCA 

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
x_pca = pca.fit_transform(x_cluster_scaled)

plt.figure(figsize=(6,5))
plt.scatter(x_pca[:,0], x_pca[:,1], c=df['Cluster'], cmap='viridis')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('Customer Segments (PCA Visualization)')
plt.colorbar(label='Cluster')
plt.show()
```

PCA 1 :'TotalSpending', 'Income'

PCA 2:  'Age', 'TotalChildren'
